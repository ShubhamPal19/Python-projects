```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

c   = (cv2.cvtColor(cv2.imread("center.jpeg"),cv2.COLOR_BGR2RGB))

tl  = cv2.cvtColor(cv2.imread("top_left.jpg"),cv2.COLOR_BGR2RGB)
tr  = cv2.cvtColor(cv2.imread("top_right.jpg"),cv2.COLOR_BGR2RGB)
br  = cv2.cvtColor(cv2.imread("bottom_right.jpg"),cv2.COLOR_BGR2RGB)
bl  = cv2.cvtColor(cv2.imread("bottom_left.jpg"),cv2.COLOR_BGR2RGB)

c= cv2.resize(c,(160,160),None)
tr= cv2.resize(tr,(320,320),None)
tl= cv2.resize(tl,(320,320),None)
br= cv2.resize(br,(320,320),None)

bl= cv2.resize(bl,(320,320),None)

bg= np.zeros((700,700,3),np.uint8)
center= np.zeros((200,200,3),np.uint8)

center[20:180,20:180]=c


bg[20:340,20:340] = tl
bg[20:340,360:680] = tr
bg[360:680,20:340] = bl 
bg[360:680,360:680] = br

bg[250:450,250:450]=center


plt.imshow(bg)

```
